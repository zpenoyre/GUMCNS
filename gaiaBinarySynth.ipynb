```python
import numpy as np
import astropy
from astropy import units as u
import astromet
import dev.astromet.astromet as astromet
import matplotlib.pyplot as plt
import matplotlib as mpl
from tqdm.notebook import tqdm
import scanninglaw.times
from scanninglaw.source import Source
```

```python
# slooow step - see scanning_law package for details
dr3_sl=scanninglaw.times.dr2_sl(version='dr3_nominal')
```

```python
dr3Period=34/12 # in years
mas=astromet.mas # conversion from degrees to milli-arcseconds
```

```python
mags=np.linspace(-5,25,100)
alerrs=astromet.sigma_ast(mags)
ax=plt.gca()
ax.plot(mags,alerrs)
ax.set_yscale('log')
ax.set_xlabel('apparent magnitude')
ax.set_ylabel('astrometric error - $\sigma_{ast}$ [mas]')
plt.show()
```

```python
nTest=16

for i in tqdm(range(nTest)):
    
    params=astromet.params()
    params.ra=360*np.random.rand()
    params.dec=np.arcsin(-1+2*np.random.rand())*180/np.pi
    
    c=Source(params.ra,params.dec,unit='deg')
    sl=dr3_sl(c, return_times=True, return_angles=True)
    ts=2010+np.squeeze(np.hstack(sl['times']))/365.25
    sort=np.argsort(ts)
    ts=ts[sort]
    
    phis=np.squeeze(np.hstack(sl['angles']))[sort]
    
    params.parallax=10*np.power(np.random.rand(),-1/3) # all within 100 pc
    params.pmrac=params.parallax*np.random.normal(scale=6.67)
    params.pmdec=params.parallax*np.random.normal(scale=6.67)
    params.period=10**(-1.5+3*np.random.rand()) # periods between 0.03 and 30 years
    params.l=np.random.rand() # uniform light ratio
    params.q=4*np.random.rand()**2 # mass ratios between 0 and 4 (half less than 1)
    params.a=10*np.random.rand()**2
    params.e=np.random.rand()
    params.vtheta=np.arccos(-1+2*np.random.rand())
    params.vphi=2*np.pi*np.random.rand()
    params.vomega=2*np.pi*np.random.rand()
    orbitalPhase=np.random.rand() # fraction of an orbit completed at t=0
    params.tperi=params.period*orbitalPhase
    
    mag=2.5+17.5*np.random.rand()
    alError=astromet.sigma_ast(mag)
    
    trueRacs,trueDecs=astromet.track(ts,params)
    
    t_obs,x_obs,phi_obs,rac_obs,dec_obs=astromet.mock_obs(ts,phis,trueRacs,trueDecs,err=alError)
    
    fitresults=astromet.fit(t_obs,x_obs,phi_obs,alError,params.ra,params.dec)
    results=astromet.gaia_results(fitresults)
    
    #optionally plot tracks
    plotts=np.sort(ts[0]+(ts[-1]-ts[0])*np.random.rand(1000))
    plotRacs,plotDecs=astromet.track(plotts,params)
    
    fig=plt.figure(figsize=(8,6))
    grid=mpl.gridspec.GridSpec(2,3)
    ax=plt.subplot(grid[0:2,0:2])
    ax.set_title('track')
    fitparams=astromet.params()
    for i in range(32):
        fitparams.ra=params.ra
        fitparams.dec=params.dec
        fitparams.drac=fitresults['drac']+np.random.randn()*fitresults['drac_error']
        fitparams.ddec=fitresults['ddec']+np.random.randn()*fitresults['ddec_error']
        fitparams.pmrac=fitresults['pmrac']+np.random.randn()*fitresults['pmrac_error']
        fitparams.pmdec=fitresults['pmdec']+np.random.randn()*fitresults['pmdec_error']
        fitparams.parallax=fitresults['parallax']+np.random.randn()*fitresults['parallax_error']
        
        fitRacs,fitDecs=astromet.track(plotts,fitparams)
        ax.plot(fitRacs,fitDecs,c='k',alpha=0.2,lw=0.5)
    ax.scatter(rac_obs,dec_obs,c=t_obs,alpha=0.2,s=50,cmap='twilight_shifted')
    ax.scatter(plotRacs,plotDecs,c=plotts,s=10,alpha=0.8,cmap='twilight_shifted')
    ax.set_xlabel(r'$\delta \ RA \cos(Dec) \ \ [mas]$')
    ax.set_ylabel(r'$\delta \ Dec \ \ [mas]$')
    ax.set_aspect('equal')
    
    ax=plt.subplot(grid[0,2])
    ax.set_title('true residual')
    comRacs,comDecs=astromet.track(plotts,params,comOnly=True)
    comRac_obs,comDec_obs=astromet.track(t_obs,params,comOnly=True)
    
    ax.scatter(rac_obs-comRac_obs,dec_obs-comDec_obs,c=t_obs,
               alpha=0.2,s=50,cmap='twilight_shifted')
    ax.scatter(plotRacs-comRacs,plotDecs-comDecs,c=plotts,
               s=10,alpha=0.8,cmap='twilight_shifted')
    ax.set_aspect('equal')
    
    ax=plt.subplot(grid[1,2])
    ax.set_title('fit residual')
    bestparams=astromet.params()
    bestparams.ra=params.ra
    bestparams.dec=params.dec
    bestparams.drac=fitresults['drac']
    bestparams.ddec=fitresults['ddec']
    bestparams.pmrac=fitresults['pmrac']
    bestparams.pmdec=fitresults['pmdec']
    bestparams.parallax=fitresults['parallax']
    bestRacs,bestDecs=astromet.track(plotts,bestparams)
    bestRac_obs,bestDec_obs=astromet.track(t_obs,bestparams)
    
    ax.scatter(rac_obs-bestRac_obs,dec_obs-bestDec_obs,c=t_obs,
               alpha=0.2,s=50,cmap='twilight_shifted')
    ax.scatter(plotRacs-bestRacs,plotDecs-bestDecs,c=plotts,
               s=10,alpha=0.8,cmap='twilight_shifted')
    ax.set_xlabel(r'$\delta \ RA \cos(Dec) \ \ [mas]$')
    ax.set_aspect('equal')
    plt.show()
    
    print(results)
```

```python
params=astromet.params()
params.ra=20#360*np.random.rand()
params.dec=20#np.arcsin(-1+2*np.random.rand())*180/np.pi

c=Source(params.ra,params.dec,unit='deg')
sl=dr3_sl(c, return_times=True, return_angles=True)
ts=2010+np.squeeze(np.hstack(sl['times']))/365.25
sort=np.argsort(ts)
ts=ts[sort]

phis=np.squeeze(np.hstack(sl['angles']))[sort]

M1=1#10**(-0.65+0.6*np.random.randn())

params.parallax=10
params.pmrac=4.5
params.pmdec=3
params.period=np.sqrt(10)
params.e=0.5
params.vtheta=np.pi/4#np.arccos(-1+2*np.random.rand())
params.vphi=np.pi/4#2*np.pi*np.random.rand()
params.vomega=-np.pi/4#2*np.pi*np.random.rand()
orbitalPhase=0.0#np.random.rand() # fraction of an orbit completed at t=0
params.tperi=params.period*orbitalPhase

qs=[0,0.8,2/M1,5/M1,10/M1]
ls=[0,0.8**3.5,0,0,0]
names=['Single Star','Stellar Binary',r'2 $M_\odot$ BH',r'5 $M_\odot$ BH',r'10 $M_\odot$ BH']

alError=0.1
```

```python
mpl.rcParams['axes.edgecolor']='w'

fig=plt.figure(figsize=(12,12))
grid=mpl.gridspec.GridSpec(5,5,width_ratios=(1,0.5,0.5,0.5,0.5))
for k in range(5):
    params.q=qs[k]
    params.l=ls[k]
    
    M2=params.q*M1
    params.a=np.power((astromet.Galt*(M1+M2)*params.period**2)/(4*np.pi**2),1/3)
    
    trueRacs,trueDecs=astromet.track(ts,params)

    t_obs,x_obs,phi_obs,rac_obs,dec_obs=astromet.mock_obs(ts,phis,trueRacs,trueDecs,err=alError)

    fitresults=astromet.fit(t_obs,x_obs,phi_obs,alError,params.ra,params.dec)
    results=astromet.gaia_results(fitresults)
    
    span2=np.flatnonzero(t_obs<astromet.tbegin+22/12)
    t_obs2=t_obs[span2]
    x_obs2=x_obs[span2]
    phi_obs2=phi_obs[span2]
    rac_obs2=rac_obs[span2]
    dec_obs2=dec_obs[span2]

    fitresults2=astromet.fit(t_obs2,x_obs2,phi_obs2,alError,params.ra,params.dec)
    results2=astromet.gaia_results(fitresults2)

    #optionally plot tracks
    plotts=np.linspace(astromet.tbegin, astromet.tbegin+34/12,1000)
    plotts2=np.linspace(astromet.tbegin, astromet.tbegin+22/12,int(1000*22/34))
    plotRacs,plotDecs=astromet.track(plotts,params)
    plotRacs2,plotDecs2=astromet.track(plotts2,params)
    
    ax=plt.subplot(grid[k,0])
    if k==0:
        ax.set_title('track')
    fitparams=astromet.params()
    for i in range(32):
        fitparams.ra=params.ra
        fitparams.dec=params.dec
        fitparams.drac=fitresults['drac']+np.random.randn()*fitresults['drac_error']
        fitparams.ddec=fitresults['ddec']+np.random.randn()*fitresults['ddec_error']
        fitparams.pmrac=fitresults['pmrac']+np.random.randn()*fitresults['pmrac_error']
        fitparams.pmdec=fitresults['pmdec']+np.random.randn()*fitresults['pmdec_error']
        fitparams.parallax=fitresults['parallax']+np.random.randn()*fitresults['parallax_error']

        fitRacs,fitDecs=astromet.track(plotts,fitparams)
        ax.plot(fitRacs,fitDecs,c='k',alpha=0.1,lw=0.5,ls='--')
    for i in range(32):
        fitparams.ra=params.ra
        fitparams.dec=params.dec
        fitparams.drac=fitresults2['drac']+np.random.randn()*fitresults2['drac_error']
        fitparams.ddec=fitresults2['ddec']+np.random.randn()*fitresults2['ddec_error']
        fitparams.pmrac=fitresults2['pmrac']+np.random.randn()*fitresults2['pmrac_error']
        fitparams.pmdec=fitresults2['pmdec']+np.random.randn()*fitresults2['pmdec_error']
        fitparams.parallax=fitresults2['parallax']+np.random.randn()*fitresults2['parallax_error']

        fitRacs,fitDecs=astromet.track(plotts2,fitparams)
        ax.plot(fitRacs,fitDecs,c='b',alpha=0.1,lw=0.5,ls='--')
    ax.scatter(rac_obs,dec_obs,c=t_obs,alpha=0.2,s=50,cmap='twilight_shifted')
    ax.scatter(plotRacs,plotDecs,c=plotts,s=10,alpha=0.8,cmap='twilight_shifted')
    if k==4:
        ax.set_xlabel(r'$\delta \ RA \cos(Dec) \ \ [mas]$')
        ax.set_ylabel(r'$\delta \ Dec \ \ [mas]$')
    ax.set_aspect('equal')

    ax=plt.subplot(grid[k,1])
    if k==0:
        ax.set_title('true residual')
    comRacs,comDecs=astromet.track(plotts,params,comOnly=True)
    comRac_obs,comDec_obs=astromet.track(t_obs,params,comOnly=True)

    ax.scatter(rac_obs-comRac_obs,dec_obs-comDec_obs,c=t_obs,
               alpha=0.2,s=50,cmap='twilight_shifted')
    ax.scatter(plotRacs-comRacs,plotDecs-comDecs,c=plotts,
               s=10,alpha=0.8,cmap='twilight_shifted')
    ax.set_aspect('equal')

    ax=plt.subplot(grid[k,2])
    if k==0:
        ax.set_title('DR2 residual')
    bestparams=astromet.params()
    bestparams.ra=params.ra
    bestparams.dec=params.dec
    bestparams.drac=fitresults2['drac']
    bestparams.ddec=fitresults2['ddec']
    bestparams.pmrac=fitresults2['pmrac']
    bestparams.pmdec=fitresults2['pmdec']
    bestparams.parallax=fitresults2['parallax']
    bestRacs,bestDecs=astromet.track(plotts2,bestparams)
    bestRac_obs,bestDec_obs=astromet.track(t_obs2,bestparams)

    cs=mpl.cm.twilight_shifted((t_obs2-t_obs[0])/(t_obs[-1]-t_obs[0]))
    ax.scatter(rac_obs2-bestRac_obs,dec_obs2-bestDec_obs,c=cs,
               alpha=0.2,s=50)
    cs=mpl.cm.twilight_shifted((plotts2-plotts[0])/(plotts[-1]-plotts[0]))
    ax.scatter(plotRacs2-bestRacs,plotDecs2-bestDecs,c=cs,
               s=10,alpha=0.8)
    #ax.set_xlabel(r'$\delta \ RA \cos(Dec) \ \ [mas]$')
    ax.set_aspect('equal')
    
    ax=plt.subplot(grid[k,3])
    if k==0:
        ax.set_title('eDR3 residual')
    bestparams=astromet.params()
    bestparams.ra=params.ra
    bestparams.dec=params.dec
    bestparams.drac=fitresults['drac']
    bestparams.ddec=fitresults['ddec']
    bestparams.pmrac=fitresults['pmrac']
    bestparams.pmdec=fitresults['pmdec']
    bestparams.parallax=fitresults['parallax']
    bestRacs,bestDecs=astromet.track(plotts,bestparams)
    bestRac_obs,bestDec_obs=astromet.track(t_obs,bestparams)

    ax.scatter(rac_obs-bestRac_obs,dec_obs-bestDec_obs,c=t_obs,
               alpha=0.2,s=50,cmap='twilight_shifted')
    ax.scatter(plotRacs-bestRacs,plotDecs-bestDecs,c=plotts,
               s=10,alpha=0.8,cmap='twilight_shifted')
    #ax.set_xlabel(r'$\delta \ RA \cos(Dec) \ \ [mas]$')
    ax.set_aspect('equal')
    
    ax=plt.subplot(grid[k,4])
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.text(0.2,0.8,names[k])
    ax.text(0.1,0.7,r'$\varpi_{DR2}=$'+)
    ax.set_axis_off()
    
    
plt.show()

mpl.rcParams['axes.edgecolor']='k'
```
